# Yearly Sales by Region

```python
import pandas as pd
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])
dataset['Year'] = dataset['Order Date'].dt.year
# Group by city and year, then calculate total sales
yearly_sales_by_region = dataset.groupby(['Region', 'Year'])['Sales'].sum().reset_index()
print(yearly_sales_by_region)
```
